```python
import sys
import os
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from selenium import webdriver
import time
import tqdm
from tqdm import tqdm_notebook
```

```python
# 크롬 웹브라우저 실행
path = "chromedriver.exe"

driver = webdriver.Chrome(path)
driver.get('http://www.naver.com')
time.sleep(2)

# 웹툰 카테고리 선택
driver.find_element_by_link_text("웹툰").click( )

# 전체 웹툰 선택
driver.find_element_by_link_text("웹툰").click( )

# 장르별 웹툰 선택
driver.find_element_by_link_text("장르별").click( )
```

```python
overlays = '//*[@id="content"]/div[3]/ul/li[1]/dl/dt/a'                         
title = driver.find_element_by_xpath(overlays)
titlename = title.text
titlename
```

```python
number_finder = '//*[@id="content"]/div[2]/h3/em'
total_number_text = driver.find_element_by_xpath(number_finder)
total_number = total_number_text.text
total_number = total_number[2:]
total_number = int(total_number)
total_number
```

```python
# 에피소드 장르 만화 정보
dict1 = {}
title_list = []

# 제목 크롤링 시작
for i in range(total_number):
    
    target_info = {}
    
    overlays1 = '.thumb > a'
    titles_raw = driver.find_elements_by_css_selector(overlays1)
    title = titles_raw[i].get_attribute('title')
    title_list.append(title)
    target_info['title'] = title
    
    overlays2 = '.desc'
    author_raw = driver.find_elements_by_css_selector(overlays2)
    author = author_raw[i].text
    target_info['author'] = author
    
    overlays3 = '.rating_type'
    star_raw = driver.find_elements_by_css_selector(overlays3)
    star = star_raw[1].text.replace('평점\n','')
    target_info['star'] = star
    
    dict1[i] = target_info
    
dict1
```
